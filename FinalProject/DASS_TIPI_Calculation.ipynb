##Calculating Extraversion Score

```python
extraversion_score=[]

for i in range(39775):
  avg=0
  a=df["TIPI1"][i]
  x=df["TIPI6"][i]
  if a==0 or x==0:
    extraversion_score.append(0)
    continue
  b=reversed_score[x]
  avg=(a+b)/2
  extraversion_score.append(avg)
```

```python
len(extraversion_score)
```

```python
df["Extraversion Score"]=extraversion_score
```
